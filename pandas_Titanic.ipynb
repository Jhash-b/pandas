Merging/Joining two dataframes

### Merge/Join two dataframes

<div><center>
<img src="https://pandas.pydata.org/docs/_images/08_merge_left.svg" width="1000"/>
</center></div>

```python
passenager_details = pd.DataFrame({
    "passenager_id": ["p0","p1","p3","p5"],
    "gender": ["male","female","male","female"],
    "age": [23,34,34,33]
})
```

```python
passenager_details.head()
```

```python
passenager_fare_details = pd.DataFrame({
    "passenager_id": ["p0","p7","p5","p3"],
    "class": ["First","Second","First","Third"],
    "fare": ["$60","$39","$80","$30"]
})
```

```python
passenager_fare_details.head()
```

```python
passenager_details.merge(passenager_fare_details, on = "passenager_id", how = "left")
```

```python
passenager_fare_details.merge(passenager_details, on = "passenager_id", how = "left")
```

```python
merged_df = passenager_details.merge(passenager_fare_details, on = "passenager_id", how = "left")
merged_df
```

```python
merged_df1 = passenager_details.merge(passenager_fare_details, on = "passenager_id", how = "right")
merged_df1
```

```python
merged_df2 = passenager_details.merge(passenager_fare_details, on = "passenager_id", how = "outer")
merged_df2
#merging both df
```

```python
merged_df3 = passenager_details.merge(passenager_fare_details, on = "passenager_id", how = "inner")
merged_df3
#it will take intersect(common) data from both df
```
